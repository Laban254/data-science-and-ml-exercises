```python
# imports
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

1. Plotting Options

```python
# Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(5, 4))
plt.plot(x, y, label='Line Plot')
plt.title('Line Plot Example')
plt.legend()
plt.show()
```

```python

# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(5, 4))
plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.show()
```

```python

# Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 9, 4]
plt.figure(figsize=(5, 4))
plt.bar(categories, values, color='green')
plt.title('Bar Chart Example')
plt.show()
```

```python

# Histogram
data = np.random.randn(1000)
plt.figure(figsize=(5, 4))
plt.hist(data, bins=30, color='purple')
plt.title('Histogram Example')
plt.show()
```

```python

# Pie Chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.figure(figsize=(5, 4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```

```python

# 3D Plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title('3D Plot Example')
plt.show()
```

2. Customize

```python
plt.figure(figsize=(6, 4))
plt.plot(x, y, color='orange', linestyle='--', linewidth=2, marker='o')
plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

3. Subplots and Multiple Figures

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.plot(x, y1, label='Sine', color='blue')
ax1.set_title('Sine Function')
ax2.plot(x, y2, label='Cosine', color='red')
ax2.set_title('Cosine Function')
plt.tight_layout()
plt.show()
```

4. Integration with Pandas and NumPy

```python
import pandas as pd

# Creating a DataFrame
df = pd.DataFrame({'X': np.linspace(0, 10, 100), 'Y': np.sin(np.linspace(0, 10, 100))})
df.plot(x='X', y='Y', kind='line', title='Plotting with Pandas', legend=True)
plt.show()
```

5. Object-Oriented and Pyplot Interfaces

```python
fig, ax = plt.subplots()
ax.plot(x, y, label='Object-Oriented Plot')
ax.set_title('Object-Oriented Interface')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
plt.show()
```

6. Interaction and Animations

```python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
x, y = [], []
line, = ax.plot([], [], color='purple')

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return line,

def update(frame):
    x.append(frame)
    y.append(np.sin(frame))
    line.set_data(x, y)
    return line,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128), init_func=init, blit=True)
plt.show()
```

11. Event Handling and Custom Callbacks

```python
def on_click(event):
    print(f'You clicked at coordinates: ({event.xdata}, {event.ydata})')

fig, ax = plt.subplots()
ax.plot(x, y)
fig.canvas.mpl_connect('button_press_event', on_click)
plt.show()
```
